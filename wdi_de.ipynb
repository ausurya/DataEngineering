```pyspark
#Creating a function to display the data frames in a pretty format
def showDF(df, limitRows =  20, truncate = True):
    if(truncate):
        pd.set_option('display.max_colwidth', 50)
    else:
        pd.set_option('display.max_colwidth', -1)
    pd.set_option('display.max_rows', limitRows)
    display(df.limit(limitRows).toPandas())
    pd.reset_option('display.max_rows')
```

# Analytic Requirement 1

### Transformation of country data

```pyspark
countryDim = country \
    .select("2-alpha code", "Country Code", "Short Name", "Long Name", "Region", "Income Group") \
    .withColumnRenamed("2-alpha code", "country_iso_code") \
    .withColumnRenamed("Country Code", "wb_country_code") \
    .withColumnRenamed("Short Name", "country_name") \
    .withColumnRenamed("Long Name", "country_long_name") \
    .withColumnRenamed("Region", "region") \
    .withColumnRenamed("Income Group", "income_group")
    
showDF(countryDim)
```

```pyspark
countryDim.printSchema()
```

```pyspark
countryDim.createOrReplaceTempView("country")
```

```pyspark
transformQuery = """
select 
    country_iso_code
    , wb_country_code
    , country_name as name
    , country_long_name as long_name
    , region
    , income_group
from 
    country
"""

#showDF(spark.sql(transformQuery))
sample = spark.sql(transformQuery)
```

```pyspark
sample.printSchema()
```

```pyspark
sample.select('name').distinct().count()
```

### Check Data quality

#### 1. Check the country codes 

```pyspark
countryCodeLengthQuery = """
select 
    length(country_iso_code) as column_length
    , count(1) as cnt
from 
    country
group by 
    length(country_iso_code)
having 
    count(1) > 1
"""
check = spark.sql(countryCodeLengthQuery)
```

```pyspark
check.show()
```

```pyspark
#It seems like there is one country with a different country code or a null code
checkOddRecord = """
select 
    country_iso_code
    ,country_name
from
    country
where
    length(country_iso_code) IS NULL
"""

showDF(spark.sql(checkOddRecord))
```

#### 2. Check duplicate records

```pyspark
countryQuery = """
select 
    country_iso_code
    , count(*) as cnt
from 
    country
group by 
    country_iso_code
having cnt > 1
"""
check = spark.sql(countryQuery)
```

```pyspark
check.show()
```

#### Data quality checks

#### 1. Duplicate indicator codes

```pyspark
dupRecQuery = """
select indicator_code,
    count(*) as count
from series
group by indicator_code
having count > 1
"""

dupRecRes = spark.sql(dupRecQuery)
showDF(dupRecRes)
```

#### 2. Are the aggregation methods case uniform?

```pyspark
checkCaseQuery = """
select distinct aggregation_method from series
"""

checkCaseRes = spark.sql(checkCaseQuery)
showDF(checkCaseRes)
```

```pyspark
# As you can see that the case of aggregate methods('Weighted average' is not same as 'Weighted Average') But we can filter this out using lower function
```

```pyspark
#Get simple aggregates 
simpleAggIndicators = seriesDimFinal \
    .filter("lower(aggregation_method) = 'sum'") \
    .select("indicator_code", "indicator_name") \
    .orderBy("indicator_code")

showDF(simpleAggIndicators, limitRows = 500, truncate = False)
```

```pyspark
# get the indicators in which we are interested 
targetIndicators = simpleAggIndicators \
    .filter("lower(indicator_name) like '%population%total%' " + 
            " or lower(indicator_name) like '%cellular%' " +
            " or lower(indicator_name) like '%broadband%'") \
    .filter("lower(indicator_name) not like '%refugee%'")

showDF(targetIndicators)
```

```pyspark
# Keep the columns that are relevant for further transformations
indicatorsData = data \
    .withColumnRenamed("Indicator Code", "indicator_code") \
    .withColumnRenamed("Country Code", "wb_country_code") \
    .drop("Indicator Name") \
    .drop("Country Name") \
    .drop("_c62")
```

```pyspark
targetIndicatorsData = indicatorsData.join(targetIndicators \
                                         , indicatorsData.indicator_code == targetIndicators.indicator_code).drop(targetIndicators.indicator_code)
```

```pyspark
indicatorsSample = targetIndicatorsData \
    .select(col("wb_country_code")
            , col("indicator_code")
            , lit("1960").alias("year")
            , col("1960").alias("indicator_value")) \
    .filter("indicator_value >= 0.0")

showDF(indicatorsSample)
```

```pyspark
targetIndicatorsData.schema.names
```

```pyspark
yearList = [x for x in targetIndicatorsData.schema.names \
           if x != 'wb_country_code' and x != 'indicator_code' and x != 'indicator_name']

print(yearList)
```

```pyspark
indicatorsDF = indicatorsSample.filter('1 = 0')
#indicatorsDF.show()

#Iterate through the list of years and store the rows in the DataFrame we created above
for indicatorYear in yearList:
    print("Processing indicators for " + indicatorYear)
    yearIndicatorDF = targetIndicatorsData \
        .select(col("wb_country_code")
                , col("indicator_code")
                , lit(indicatorYear).alias("year")
                , col(indicatorYear).alias("indicator_value")) \
        .filter("indicator_value >= 0")
    indicatorsDF = indicatorsDF.union(yearIndicatorDF)
    
indicatorsDF.show()    
```

```pyspark
#cache the data frame because we are going to iterate this frequently in further processing
indicatorsDF.cache()
indicatorsDF.count()
```

```pyspark
#convert the column from string to numeric[this is what happens when you don't use inferschema]
indicatorsDF = indicatorsDF.withColumn("indicator_value", indicatorsDF["indicator_value"].cast(IntegerType()))
```

```pyspark
indicatorsDF.printSchema()
```

```pyspark
yearPivot = indicatorsDF.groupBy('year').pivot('indicator_code').sum('indicator_value')
```

```pyspark
yearPivot.orderBy('year').show()
```

```pyspark
yearPivotDF = yearPivot.orderBy('year') \
    .withColumnRenamed('IT.CEL.SETS', 'cellular_subscriptions') \
    .withColumnRenamed('IT.NET.BBND', 'broadband_subscriptions') \
    .withColumnRenamed('SP.POP.0014.TO', 'population_age_0_to_14') \
    .withColumnRenamed('SP.POP.1564.TO', 'population_age_15_64') \
    .withColumnRenamed('SP.POP.65UP.TO', 'population_age_65_and_above') \
    .withColumnRenamed('SP.POP.TOTL', 'population')
```

```pyspark
yearPivotDF.show()
```

```pyspark
# Caching the yearPivotDF so that we can iterate over it for data quality checks
yearPivotDF.cache()
yearPivotDF.count()
```
